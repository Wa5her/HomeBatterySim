```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
df = pd.read_csv('output.csv',parse_dates=['interval_start','interval_end'])
```

```python
# Convert interval_start to datetime objects
df['interval_start'] = pd.to_datetime(df['interval_start'],utc=True)

# Extract the time portion and format it as HH:MM
df['time'] = df['interval_start'].dt.strftime('%H:%M')
```

```python
# Create the box plot using Plotly
fig = go.Figure()

metric = 'sell_rate_inc_vat'
fig.add_trace(go.Box(
    x=df['time'],
    y=df[metric],
    name=metric,
    marker_color='green'
))
metric = 'buy_rate_inc_vat'
fig.add_trace(go.Box(
    x=df['time'],
    y=df[metric],
    name=metric,
    marker_color='red'
))

# Customize the layout
fig.update_layout(
    title=f'{metric} per 30-Minute Interval (4 PM - 4 PM)',
    xaxis_title='Time of Day',
    yaxis_title=metric,
    xaxis=dict(
        categoryorder='array',
        categoryarray=sorted(df['time'].unique())  # Ensure correct order on x-axis
    )
    
)

fig.show()
```

```python
#Trade strategy charge from 11.30 pm to 4.30 AM for 5 hours  = 40 kWh
def buy_time(row):
    if row['time'] >= '23:30' and row['time'] <= '04:30':
        return -8*0.5*row['buy_rate_inc_vat'] # 8 kWh per hour charge
    elif row['time'] >= '15:00' and row['time'] <= '16:30':
        return 0
```
